### 2. Exploring Relationships

```python
# Create a correlation matrix using pandas
df.corr(numeric_only=True)
```

### Correlations

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df.corr(numeric_only=True), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap
```

The correlation between employee_id and age, length of service is a given . 

There is  a moderate correlation  between store number and population.

### Scatterplots

```python
# length_of_service is more closely correlated with age rather than job category
# Create a scatterplot for the "age" and "length_of_service" columns in seaborn
sns.lmplot(x = 'age', y = 'length_of_service', data = df)
```

The largest service peak of around 13 years overlaps with the age peak of above 60 years and with ages between 40-50.They are people who changed careers to join the company.

### Pair Plots

```python
# Keep only the variables you want to use in the pair plot

sub=df[[ 'age','year_of_record','length_of_service','population' ]]
```

```python
# Create a pair plot 

g= sns.pairplot(sub)
```

### Categorial Plot

```python
# Create a histogram
sns.histplot(sub['length_of_service'], bins = 20, kde = True)
```

```python
# Create a histogram
sns.histplot(sub['age'], bins = 20, kde = True)
```

```python
# Create a categorical plot in seaborn 

sns.set(style="ticks")
g = sns.catplot(x="age", y="job_category", data=df)
```

### 3.Visualizations for exploring relationships

```python
#bar chart to show gender comparison
df['gender'].value_counts().plot.bar()
```

```python
#bar chart to show employment status comparison
df['employment_status'].value_counts().plot.bar()
```

```python
#bar chart to show population according to city 
df['population_flag'].value_counts().plot.bar()
```

```python
#bar chart to show job category comparison
df['job_category'].value_counts().plot.bar()
```

```python
plt.figure(figsize=(5, 5))
sns.countplot(x="age_flag", hue="employment_status", data=df)
```

#### Terminations are more after the 60 years of age(Old-age)

```python
plt.figure(figsize=(5, 5))
sns.countplot(x="population_flag", hue="employment_status", data=df)
```

#### Maximum number of employees are from the mega city in which the ratio of active to terminated status of the employee is higher.
#### Number of employees from the town are less compared to mega city but much higher compared to rural area. 
#### The least number of employees are from rural areas and the ratio of active to terminated status of the employee is lower.  

```python
#bar chart to show city name comparison
df['city_name'].value_counts().plot.bar()
```

#### In the above bar chart, we see that Vancouver and Victoria have the highest share of terminated employees. This skewness might be because the higest concentration of employees in the dataset is in these 2 cities. 

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="job_category", hue="employment_status", data=df)
```

Maximum number of "Employees" are the ones with the employee job title and the number of ones with active status is more than the number of ones with terminated status.

```python
plt.figure(figsize=(5, 5))
sns.countplot(x="year_of_record", hue="employment_status", data=df)
```

2014 had an unusually high number of employment terminations.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x="business_unit", hue="employment_status", data=df)
```

####  More number of employees are from the stores and not from the head-office and cases of termination are also higher in stores  than cases of termination in the head-office .
